```python
# import statements and configuration
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# set theme
sns.set_theme(style="darkgrid")
figure_size=[10, 5]
```

```python
# read data from csv to df
data = pd.read_csv('../../data/raw/world-happiness-report-2021.csv')
data.head()
```

```python
# check for missing value
data.isnull().sum() # no missing value :D
```

```python
# removing unwanted cols
df = data[data.columns[:-7]]
```

```python
# plotting boxplot per region
fig, ax = plt.subplots(figsize=figure_size)

sns.boxplot(y='Regional indicator', x='Ladder score', data=df, palette='coolwarm', ax=ax)
sns.stripplot(y='Regional indicator', x='Ladder score', data=df, color='#114b5f', size=5, alpha=.7, ax=ax)

ax.set(title='Boxplot by Regional Indicator', ylabel='')
plt.show()
```
